```python
import pandas as pd
```

```python
data = pd.read_csv('../data/combined_data.csv')
```

```python
# Available columns: Adj Close, Close, High, Low, Open, Volume

# Drop row 1 

data = data.drop(data.index[1])
```

```python
# Extract tickers (row 1)

tickers = data.iloc[0, 1:].values

data_copy = data.copy()

# Merge tickers and columns starting from the second column

# For each column make it "previous column name + ticker"

for i in range(1, len(data.columns)):
    data_copy.iloc[0, i] = data.columns[i] + ' ' + tickers[i-1]
```

```python
# Drop the original columns and make the first row the new column names

data_copy.columns = data_copy.iloc[0]
```

```python
# Drop the first row

data_copy = data_copy.drop(data_copy.index[0])
```

```python
# Rename Ticker to Date

data_copy = data_copy.rename(columns={'Ticker': 'Date'})
```

```python
# Convert all columns to numeric except Date

for col in data_copy.columns:
    if col != 'Date':
        data_copy[col] = pd.to_numeric(data_copy[col], errors='coerce')
```

```python
# Linearly interpolate missing values except for the Date column

data_copy.iloc[:, 1:] = data_copy.iloc[:, 1:].interpolate(method='linear', axis=1)
```

```python
import re

data_copy.columns = [re.sub(r'\.\d+', '', col) for col in data_copy.columns]
```

```python
data_copy.reset_index(drop=True, inplace=True)
```

```python
data_copy.to_csv('../data/clean_data.csv', index=False)
```
